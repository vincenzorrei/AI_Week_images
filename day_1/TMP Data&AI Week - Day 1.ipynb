## 2.2) Tipi

### üî¢ `int` ‚Äì Numeri interi


Un tipo di dato che rappresenta **numeri interi**, positivi o negativi, **senza parte decimale**.

Esempi: `5`, `-2`, `0`, `123456`

````python
a = 123
print("Il valore √®:", a)
print("Tipo:", type(a))
````

````python
# Somma
x = 1
y = 2
somma = x + y
print("La somma √®:", somma)
````

### üåä `float` ‚Äì Numeri decimali


Il tipo `float` rappresenta **numeri con la virgola**, utili per misure, percentuali, e calcoli pi√π precisi.

Esempi: `3.14`, `-0.5`, `2.0`

````python
pi = 3.14
print("Il valore √®:", pi)
print("Tipo:", type(pi))
````

````python
# Operazioni matematiche (moltiplicazione)
prezzo = 3.5
quantit√† = 5
totale = prezzo * quantit√†
print("Totale:", totale)
````

### üìù `str` ‚Äì Stringhe di testo


Le stringhe (`str`) rappresentano **sequenze di caratteri**, cio√® **testo**. Si scrivono tra **virgolette**.

Esempi: `"ciao"`, `'Python'`, `"123"`

````python
messaggio = "Ciao a tutti"
print(messaggio)
print("Tipo:", type(messaggio))
````

````python
# Concatenazione
nome = "Alice"
saluto = "Ciao, " + nome + "!" # "Ciao, Alice!"
print(saluto)
````

### üîò `bool` ‚Äì Valori booleani


Un tipo di dato logico con soli due possibili valori: **`True` (vero)** o **`False` (falso)**. Molto usato in **condizioni e controlli**.

Esempi: `True`, `False`

````python
attivo = True
print("Stato:", attivo)
print("Tipo:", type(attivo))
````

````python
# Check
et√† = 20
maggiorenne = et√† == 18
print("Sei maggiorenne?", maggiorenne)
````

### üìö `list` ‚Äì Collezioni ordinate di elementi


Una `list` √® una **sequenza ordinata** di elementi, che possono essere di qualsiasi tipo. Si definisce con le **parentesi quadre `[]`**.

√à **modificabile** (mutable): si possono aggiungere, rimuovere o cambiare elementi.

Esempi: `[1, 2, 3]`, `["a", "b", "c"]`

````python
frutti = ["mela", "banana", "kiwi"] # lista con mela banana e kiwi
print(frutti)
print("Tipo:", type(frutti))
````

````python
# Aggiungere elementi
numeri = [10, 20, 30]
numeri.append(40) # aggiungere elementi
print("Lista aggiornata:", numeri)
````

````python
# Accedere ad elementi
primo_numero = numeri[0]
print("Primo elemento:", primo_numero)
print("Secondo elemento:", numeri[1])
````

````python
# Rimuovere elementi
numeri.remove(30) # rimuovere elementi
print("Lista aggiornata:", numeri)
````

````python
# Check presenza di elementi
presente = 'mela' in frutti # la mela √® negli elementi?
print(presente)
````

````python
# Lista di liste
bibite = ['acqua', 'birra', 'vino']
spesa = [frutti, bibite]
print(spesa)
````

````python
# Estensione di liste
altra_spesa = frutti + bibite
print(altra_spesa)
````

````python
# Importazione libreria
import random

# Campionamento
elemento = random.choice(altra_spesa) # random choice
print(elemento)
````
